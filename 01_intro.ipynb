**What are Data Types?**
- Categories of data that Python understands
- Different types are used for different purposes
- Python recognizes the type automatically


**Integer (int)**
- Whole numbers without decimals
- Examples: `age = 25`, `count = -5`, `year = 2025`
- Use for: counting, age, quantities


**Float (float)**
- Numbers with decimals
- Examples: `height = 1.75`, `price = 9.99`, `pi = 3.14`
- Use for: measurements, prices, scientific data


**String (str)**
- Text and characters
- Use single or double quotes: `name = "Ali"` or `name = 'Ali'`
- Examples: `country = "Afghanistan"`, `message = "Hello"`
- Use for: names, addresses, messages


**Boolean (bool)**
- True or False values only
- Examples: `is_student = True`, `is_active = False`
- Use for: conditions, yes/no decisions


**How to Check Type**
- Use `type()` function: `type(age)` shows `<class 'int'>`

<div style="padding: 40px; font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; font-size: 1.1em; line-height: 1.8; color: #333;">

**What are Arithmetic Operators?**
- Symbols used to perform mathematical calculations
- Work with numbers (integers and floats)


**Common Arithmetic Operators**

- `+` **Addition**: Add two numbers
  - Example: `5 + 3 = 8`
  - Example: `10.5 + 2.5 = 13.0`

- `-` **Subtraction**: Subtract second number from first
  - Example: `10 - 4 = 6`
  - Example: `7.5 - 2.5 = 5.0`

- `*` **Multiplication**: Multiply two numbers
  - Example: `4 * 5 = 20`
  - Example: `3.5 * 2 = 7.0`

- `/` **Division**: Divide first number by second
  - Example: `10 / 2 = 5.0`
  - Example: `15 / 3 = 5.0`

- `//` **Floor Division**: Divide and round down (no decimals)
  - Example: `7 // 2 = 3`
  - Example: `10 // 3 = 3`

- `%` **Modulus**: Get the remainder after division
  - Example: `7 % 2 = 1` (7 divided by 2 leaves remainder 1)
  - Example: `10 % 3 = 1`

- `**` **Power**: Raise to a power
  - Example: `2 ** 3 = 8` (2 to the power of 3)
  - Example: `5 ** 2 = 25`

<div style="padding: 40px; font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; font-size: 1.1em; line-height: 1.8; color: #333;">

**What are Logical Operators?**
- Used to combine conditions
- Help make complex decisions
- Return True or False


**and Operator**
- Both conditions must be True to return True
- Example: `age > 18 and age < 65`
  - If age is 25: `True and True = True`
  - If age is 70: `True and False = False`
- Use: Check multiple conditions at once


**or Operator**
- At least one condition must be True to return True
- Example: `is_student or is_teacher`
  - If is_student is True: `True or False = True`
  - If both are False: `False or False = False`
- Use: When any one condition is enough


**not Operator**
- Reverses (flips) the result
- Example: `not is_active`
  - If is_active is True: `not True = False`
  - If is_active is False: `not False = True`
- Use: Opposite of a condition


**Real Examples**
- `age > 13 and age < 18` â†’ Check if someone is a teenager
- `has_license or has_passport` â†’ Check if you have ID
- `not is_raining` â†’ Check if it's not raining

<div style="padding: 40px; font-family: -apple-system, BlinkMacSystemFont, 'Segoe UI', sans-serif; font-size: 1.1em; line-height: 1.8; color: #333;">

**What are Comparison Operators?**
- Used to compare two values
- Always return True or False


**Common Comparison Operators**

- `==` **Equal to**: Check if two values are the same
  - Example: `5 == 5` returns `True`
  - Example: `5 == 3` returns `False`

- `!=` **Not equal to**: Check if two values are different
  - Example: `5 != 3` returns `True`
  - Example: `5 != 5` returns `False`

- `>` **Greater than**: Check if first value is bigger
  - Example: `10 > 5` returns `True`
  - Example: `3 > 8` returns `False`

- `<` **Less than**: Check if first value is smaller
  - Example: `3 < 8` returns `True`
  - Example: `10 < 5` returns `False`

- `>=` **Greater than or equal**: First value is bigger or same
  - Example: `5 >= 5` returns `True`
  - Example: `10 >= 8` returns `True`

- `<=` **Less than or equal**: First value is smaller or same
  - Example: `5 <= 5` returns `True`
  - Example: `3 <= 8` returns `True`
